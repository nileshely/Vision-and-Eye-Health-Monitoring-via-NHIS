# Report on NHIS Vision and Eye Health Surveillance Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

The NHIS Vision and Eye Health Surveillance Dataset provides a comprehensive overview of vision and eye health indicators collected through the National Health Interview Survey (NHIS) conducted annually by the National Center for Health Statistics at the CDC. This dataset encompasses a wide range of demographic factors including age, gender, race/ethnicity, and risk factors, allowing for detailed analysis of trends and disparities in visual function across various population groups.

```python
df = pd.read_csv("/kaggle/input/nhis-vision-and-eye-health-surveillance-dataset/NHIS_Vision_and_Eye_Health_Surveillance_20240501.csv")
```

```python
df.head()
```

```python
df.describe()
```

Upon analyzing the dataset, it is evident that gender distribution within the sample population is relatively balanced, as illustrated by the count plot of gender. Additionally, the dataset reveals insights into the age distribution, with a notable proportion of individuals falling within the middle age groups. The bar plot of age distribution highlights the distribution of sample sizes across different age groups, providing valuable information for targeted interventions and healthcare planning.

```python
# Set brown color palette
sns.set_palette(sns.color_palette(["#8B4513", "#A0522D", "#CD853F", "#D2691E", "#8B7E66", "#A67B5B", "#8B5A2B"]))
```

Exploring the relationship between data values and demographic factors such as gender and age yields further insights. For instance, box plots and violin plots demonstrate variations in data values across different gender and age groups, indicating potential disparities in visual function. Moreover, the correlation heatmap reveals associations between numerical variables, facilitating a deeper understanding of factors influencing vision and eye health outcomes.

```python
# Visualization 1: Count plot of Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Box plot of Data Value by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Data_Value')
plt.title('Data Value by Gender')
plt.xlabel('Gender')
plt.ylabel('Data Value')
plt.show()
```

```python
# Visualization 3: Violin plot of Data Value by Age
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Age', y='Data_Value')
plt.title('Data Value by Age')
plt.xlabel('Age Group')
plt.ylabel('Data Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 4: Scatter plot of Sample Size vs Data Value
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sample_Size', y='Data_Value')
plt.title('Sample Size vs Data Value')
plt.xlabel('Sample Size')
plt.ylabel('Data Value')
plt.show()
```

```python
# Visualization 5: Pie chart of Race/Ethnicity distribution
plt.figure(figsize=(10, 6))
df['RaceEthnicity'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette())
plt.title('Race/Ethnicity Distribution')
plt.ylabel('')
plt.show()
```

```python
# Visualization 6: Box plot of Data Value by Age and Gender
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Age', y='Data_Value', hue='Gender')
plt.title('Data Value by Age and Gender')
plt.xlabel('Age Group')
plt.ylabel('Data Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 7: Histogram of Data Value
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Data_Value', bins=20)
plt.title('Distribution of Data Value')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.show()
```
